```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time

statements = []  # List to store statements
values = []  # List to store values

def scrape_website(page_number):
    URL = f'https://www.politifact.com/factchecks/list/?category=coronavirus&page={page_number}'
    print(f"Scraping URL: {URL}")  # Debugging URL
    try:
        webpage = requests.get(URL)
        soup = BeautifulSoup(webpage.text, 'html.parser')

        statement_quotes = soup.find_all('div', attrs={'class': 'm-statement__quote'})
        statement_values = soup.find_all('div', attrs={'class': 'm-statement__meter'})

        print(f"Page {page_number}: Found {len(statement_quotes)} statement quotes")
        print(f"Page {page_number}: Found {len(statement_values)} statement values")

        for i in range(len(statement_quotes)):
            statement_text = statement_quotes[i].find('a').text.strip()

            if i < len(statement_values):
                img_tag = statement_values[i].find('div', attrs={'class': 'c-image'}).find('img')
                value_text = img_tag.get('alt') if img_tag else None
            else:
                value_text = None  # If there's no corresponding value, set it as None

            # Debugging to check the data
            print(f"Statement: {statement_text}, Value: {value_text}")

            statements.append(statement_text)
            values.append(value_text)

    except Exception as e:
        print(f"An error occurred on page {page_number}: {e}")

# Loop through multiple pages
for i in range(1, 101):  # Adjust this range based on actual number of pages available
    scrape_website(i)
    time.sleep(1)  # Add a delay to avoid getting blocked

# Create a DataFrame with the results
data = pd.DataFrame({'statement': statements, 'value': values})

# Save the data to a CSV file
data.to_csv('scraped_data.csv', index=False)

# Display the dataframe
print(data)

# Check the CSV file
print(data['value'].value_counts())
print(f"Total unique values: {data['value'].nunique()}")


```
